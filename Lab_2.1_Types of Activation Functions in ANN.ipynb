```python
x1 = np.arange(1,11,1)
y1 = x1
y2 = x1**2

plt.plot(y1, marker="*")
plt.plot(y2, marker=".")
plt.title("Matplot")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend(["linear", "quadratic"], loc="upper left")
plt.show()
```
